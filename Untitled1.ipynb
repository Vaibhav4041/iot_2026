```python
class parent():
    def singing(self):
        return 'I am a parent class with singing skills'
```

```python
class child(parent):
    def cricke(self):
        return 'I am a child class with cricke skill'
```

```python
c = child()
```

```python
c.cricke()
```

```python
c.singing()
```

```python
from abc import ABC, abstractmethod
```

```python
class phone(ABC):
    @abstractmethod
    def calling(self):
        pass

class iphone(phone):
    def calling(self):
        return 'I phone is going to make a call'
```

```python
p = iphone()
```

```python
p.calling()
```

```python
class samsung(phone):
    def camera(self):
        return 'Camera is opning'
```

```python
s = samsung()
```

```python
class samsung(phone):
    def calling(self):
        return 'samsung is going to make a call'
    def camera(self):
        return 'Camera is opning'
```
