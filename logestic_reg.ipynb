```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sagnikghoshcr7/Bank-Customer-Churn-Prediction/refs/heads/master/Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.drop(columns=['RowNumber',	'CustomerId',	'Surname'],inplace=True)
```

```python
df.head()
```

```python
gen_map = {'Female':0, 'Male':1}
```

```python
df['Gender'] = df['Gender'].map(gen_map)
```

```python
df['Geography'].unique()
```

```python
geo_map = {'France':0, 'Spain':1, 'Germany':2}
```

```python
df['Geography'] = df['Geography'].map(geo_map)
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
```

```python
X = df.iloc[:,:-1]
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
accuracy_score((y_test,y_pred)
```
