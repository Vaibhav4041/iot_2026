```python
import numpy as np
from tensorflow.keras.models import Sequential # archit
from tensorflow.keras.layers import Dense  # Neuron
```

```python
X = np.array([1,2,3,4,5], dtype='float')
```

```python
X
```

```python
y = np.array([20,40,60,80,100], dtype=float)
```

```python
model = Sequential()
```

```python
# model.add(Dense(3, activation='relu', input_shape=(1,) ))
model.add(Dense(50, activation='relu', input_shape=(1,) ))
```

```python
model.add(Dense(20, activation='relu'))
```

```python
model.add(Dense(1, activation='linear')) # output layer
```

```python
model.compile(optimizer='adam', loss = 'mse')
```

```python
model.fit(X,y, epochs = 500, verbose=True)
```
