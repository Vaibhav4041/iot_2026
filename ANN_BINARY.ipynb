```python
import numpy as np
from tensorflow.keras.models import Sequential # archit
from tensorflow.keras.layers import Dense  # Neuron
```

```python
X = np.array([8,6,9,6,4,1,7,2,3], dtype='float')
```

```python
y = np.array([1,1,1,1,0,0,1,0,0], dtype=float)
```

```python
model = Sequential()
```

```python
model.add(Dense(5, activation='sigmoid', input_shape=(1,) ))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy',metrics=["accuracy"] )
```

```python
history = model.fit(X,y, epochs = 50, verbose=True)
```

```python
history.history
```
